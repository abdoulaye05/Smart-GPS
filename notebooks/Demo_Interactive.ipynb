## 2Ô∏è‚É£ Cr√©ation d'un Graphe

Cr√©ons un graphe en grille 10√ó10 (100 sommets)


```python
# G√©n√©rer un graphe en grille
graph = generate_grid_graph(10, 10, spacing=1.0)

print(f"üìä Graphe cr√©√© :")
print(f"   Sommets : {graph.num_vertices()}")
print(f"   Ar√™tes : {graph.num_edges_count()}")
```

## 3Ô∏è‚É£ Algorithme de Dijkstra

**Complexit√©** : O((n+m) log n)

Trouvons le plus court chemin de 0 √† 99


```python
# Ex√©cuter Dijkstra
result_dijkstra = dijkstra(graph, 0, 99)

print(f"üéØ R√©sultats Dijkstra :")
print(f"   Co√ªt : {result_dijkstra.cost:.2f} unit√©s")
print(f"   Sommets visit√©s : {result_dijkstra.visited_nodes}")
print(f"   Temps : {result_dijkstra.execution_time*1000:.2f} ms")
```

## 4Ô∏è‚É£ Algorithme A*

**Heuristique** : Distance euclidienne


```python
# Ex√©cuter A*
result_astar = astar(graph, 0, 99)

print(f"üéØ R√©sultats A* :")
print(f"   Co√ªt : {result_astar.cost:.2f} unit√©s")
print(f"   Sommets visit√©s : {result_astar.visited_nodes}")
print(f"   Temps : {result_astar.execution_time*1000:.2f} ms")
```

## 5Ô∏è‚É£ Comparaison

‚ú® **A* est plus efficace gr√¢ce √† son heuristique !**


```python
# Comparaison
gain = (1 - result_astar.visited_nodes / result_dijkstra.visited_nodes) * 100

print(f"üìä Comparaison :")
print(f"   Dijkstra : {result_dijkstra.visited_nodes} sommets visit√©s")
print(f"   A*       : {result_astar.visited_nodes} sommets visit√©s")
print(f"   Gain     : {gain:.1f}% de r√©duction ‚ú®")
```
